## 순회 종류: 행, 열, 지그재그, 델타

전치행렬 주의: 모든 좌표에 대해서 스왑하면 그냥 그대로가 된다.

### 부분집합의 합 문제: 모든 부분집합 생성 -> 2의 n승
    - loop 생성해서 확인: 0,1: 0-보여주지 않음, 1-보여줌
    - 1<<n : 2의 n승
    - i&(1<<j) : i의 j번째 비트가 1인지 아닌지 리턴해서 확인 가능

## 검색
### 순차검색, 이진검색, 인덱싱

- 순차 검색
    - 걍 순서대로 검색하는 방법으로 비효율적일 수 있음
    - array, 연결리스트 등 순차구조로 되어있는 자료구조 사용하는게 유리
    - 정렬된 경우, 정렬되지 않은 경우
    - 정렬된 경우는 제일 큰 수, 제일 작은 수를 비교해서 범위를 아예 넘겨버리면 애초에 포기할 수 있으므로 더 좋음
    - O(n)

- 이진 검색
    - 정렬되어 있어야 함. 항상 정렬 필요
    - 효율적
    - 중간에 있는 애랑 비교
    
- 인덱스
    - 테이블 동작 속도 높여줌
    - 키 필드로 검색, 정렬
    - 대량의 데이터 정렬할때 느려질까봐 배열 사용..?

### 셀렉션 알고리즘
- 셀렉션 알고리즘
    - k번째에 있는 원소 찾기
    - k가 작을 때 유리 O(kn)

- 선택 정렬
    - 가장 작은 값의 원소부터 차례대로 선택해서 위치를 교환하는 방식
    - 최소값 find, 리스트의 맨 앞과 교환 -> 반복
    - O(n^2)
