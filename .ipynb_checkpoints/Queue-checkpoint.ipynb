# Queue

- 뒤 삽입, 앞 삭제
- First in First out
- front=-1, end=-1

### 선형 Queue
- 초기: front=rear=-1
- 공백: front=rear
- 포화: reaer=n-1

- 문제점: 포화로 잘못 인식할 수 있음 -> 원형 queue

### 원형 Queue
- front는 항상 공백으로 놔둠
- mod로 구분
- front, rear 0으로 초기화
- rear = (rear+1)mod n, front = (front+1)mod n
- 삭제할때 front가 움직이고, 삽입할때 rear가 움직인다
- 공백상태: front=rear, 포화상태: (rear+1) mod n = front

우선순위 큐, heap 자료구조
